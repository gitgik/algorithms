```python
def missing_lowest(array) -> int:
    if len(array) == 0:
        return 1
    
    while len(array) > 0:
        smallest = array.pop(array.index(min(array)))
        if smallest < 0:
            continue
        elif smallest + 1 in array:
            continue
        else:
            return smallest + 1
    return 1
```

```python
missing_lowest([-1, 0, 4, 5])
```
